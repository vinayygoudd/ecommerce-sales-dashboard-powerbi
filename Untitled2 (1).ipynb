```python
import pandas as pd
import numpy as np
df = pd.read_csv("Amazon.csv")
df.head()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
num_cols = ['Quantity', 'UnitPrice', 'Discount', 'Tax', 'ShippingCost', 'TotalAmount']
df[num_cols] = df[num_cols].fillna(0)
```

```python
text_cols = ['CustomerName', 'ProductName', 'Category', 'Brand', 'PaymentMethod', 'City', 'State']
df[text_cols] = df[text_cols].fillna("Unknown")
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
df['DiscountAmount'] = df['Revenue'] * df['Discount']
df['FinalAmount'] = df['Revenue'] - df['DiscountAmount'] + df['Tax'] + df['ShippingCost']

df['Profit'] = df['Revenue'] - (df['Tax'] + df['ShippingCost'] + df['DiscountAmount'])
df['ProfitMargin'] = df['Profit'] / df['Revenue']
```

```python
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['MonthName'] = df['OrderDate'].dt.month_name()
df['DayOfWeek'] = df['OrderDate'].dt.day_name()
df['Quarter'] = df['OrderDate'].dt.quarter
```

```python
#Visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")

sns.set_palette("deep")
```

```python
#Monthly Revenue Line Chart
monthly = df.groupby(['Year','Month'])['FinalAmount'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=monthly, x='Month', y='FinalAmount', hue='Year', marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
#Category-wise Revenue Bar Chart
category_sales = df.groupby('Category')['FinalAmount'].sum().sort_values()

plt.figure(figsize=(10,6))
category_sales.plot(kind='barh')
plt.title("Revenue by Category")
plt.xlabel("Revenue")
plt.ylabel("Category")
plt.show()
```

```python
#Top 10 Products by Quantity Sold
top_products = df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_products.plot(kind='bar')
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=75)
plt.show()
```

```python
#State-wise Revenue Bar Chart
state_sales = df.groupby('State')['FinalAmount'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
state_sales.plot(kind='bar')
plt.title("Revenue by State")
plt.ylabel("Revenue")
plt.xticks(rotation=75)
plt.show()
```

```python
#Payment Method Distribution (Pie Chart)
payment_counts = df['PaymentMethod'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Payment Method Share")
plt.show()
```

```python
#Profit vs Discount Scatter Plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title("Discount vs Profit")
plt.xlabel("Discount (%)")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
#Hourly Sales Trend
hourly = df.groupby('Hour')['FinalAmount'].sum()

plt.figure(figsize=(12,6))
sns.lineplot(x=hourly.index, y=hourly.values, marker='o')
plt.title("Hourly Sales Trend")
plt.xlabel("Hour of Day")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
#Profit Margin Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['ProfitMargin'], bins=30, kde=True)
plt.title("Profit Margin Distribution")
plt.xlabel("Profit Margin")
plt.ylabel("Frequency")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df[['Quantity','UnitPrice','Discount','Tax','ShippingCost','FinalAmount','Profit']].corr(),
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#Yearly Revenue Bar Chart
yearly = df.groupby('Year')['FinalAmount'].sum()

plt.figure(figsize=(8,6))
sns.barplot(x=yearly.index, y=yearly.values)
plt.title("Yearly Revenue Overview")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```

```python
#Pareto Chart (80/20 Rule)
pareto = df.groupby('ProductName')['FinalAmount'].sum().sort_values(ascending=False)
cum_percentage = pareto.cumsum() / pareto.sum()

plt.figure(figsize=(12,6))
ax = pareto.plot(kind='bar', color='skyblue')
plt.ylabel('Revenue')

ax2 = ax.twinx()
ax2.plot(cum_percentage.values, color='red', marker='o')
ax2.axhline(0.80, color='green', linestyle='--')
ax2.set_ylabel('Cumulative Percentage')

plt.title("Pareto Analysis (80/20 Rule)")
plt.show()
```
